```r
#Vectorize approach of multiplying 2 Vectors
N <- 100
a <- rnorm(N)
b <- rnorm(N)

c <- a*b

#De-vectorized approach of multiplying 2 Vectors
d <- rep("NA",N)
for(i in 1:N)
    {
    d[i] <- a[i] * b[i]
}
#Here we saw both give same result but for larger values of N second approach takes more time as compared to first one
print(c)
print(d)
```
